###Sıfırdan Numpy Array Oluşturma

**ones() metodu:** Birlerden oluşan dizi oluşturmak için kullanılır.

```python
a=np.ones((3,3),dtype=int)
print(a)
```

###Array Birleştirme(Concateration)

Array birleştirme (Concatenation), iki veya daha fazla diziyi (array) belirli bir eksen boyunca birleştirme işlemidir. NumPy kütüphanesinde bu işlem için np.concatenate() fonksiyonu kullanılır.

- Dikey (axis=0) birleştirme: Satır bazında birleştirme yapar. (alt alta ekler)

- Yatay (axis=1) birleştirme: Sütun bazında birleştirme yapar. (yan yana ekler)


Birleştirilecek diziler metot içinde köşeli parantezlerle verilmelidir ->
concatenate([dizi1,dizi2])

```python
# Tek boyutlu arraylerde
x=np.array([1,2,3])
y=np.array([4,5,6])

a=np.concatenate([x,y]) # diziler köşeli parantez içine yazılmalı!
print(a)
```

```python
# İki boyutlu arraylerde
a=np.arange(0,4).reshape((2,2))
b=np.arange(4,8).reshape((2,2))

# Satır bazında birleştirme(alt alta)
c=np.concatenate([a,b],axis=0)
print(c)
print("-----------------------")
# Sütun bazında birleştirme(yan yana)
d=np.concatenate([a,b],axis=1)
print(d)
```

```python
# Dikey birleştirme (satır bazlı)
combined_v = np.vstack((a, b))
print(combined_v)
print()

# Yatay birleştirme (sütun bazlı)
combined_h = np.hstack((a, b))
print(combined_h)
```

###Array Ayırma(Splitting)

Array ayırma (Splitting), büyük bir diziyi belirli parçalara bölme işlemidir. NumPy kütüphanesinde bu işlem için numpy.split() fonksiyonu kullanılır.

- Dikey (axis=0) ayırma: Diziyi satır bazında böler.

- Yatay (axis=1) ayırma: Diziyi sütun bazında böler.


```python
x=np.array([1,2,3,99,99,3,2,1])
```

```python
a,b,c=np.split(x,[3,5]) # x dizisini 0-3'e kadar, 3'ten 5. indise kadar ve 5.indisten son elemana kadar 3 parçaya ayırır!(axis=0 varsayılan değer)
print(a,b,c)
```

```python
a,b=np.split(x,[4]) # Tek bir indis verilirse 2 parçaya ayırır
print(a,b)
```

```python
# İki boyutlu array ayırma
m=np.arange(16).reshape((4,4))
print("m:",m)
print()

# vsplit() metodu satır bazında böler.
ust,alt=np.vsplit(m,[2])
print("ust:",ust)
print()
print("alt:",alt)
print()

# hsplit() metodu sütun bazında böler.
sol,sag=np.hsplit(m,[2])
print("sol:",sol)
print()
print("sag:",sag)
print()
```

###Array Sıralama(Sorting)

Array Sıralama (Sorting), bir dizideki elemanları belirli bir düzene göre sıralama işlemidir. NumPy kütüphanesinde bu işlem için numpy.sort() ve numpy.argsort() fonksiyonları kullanılır.

- numpy.sort(): Diziyi sıralar (orijinal diziyi değiştirmez!)

- numpy.argsort(): Sıralı dizinin indekslerini döndürür.

```python
a=np.array([45,56,8,9,12])
print(np.sort(a)) # Array üzerinde herhangibir değişikliğe sebep olmaz
```

```python
# Kalıcı hale getirmek için atama yapılmalı
b=np.sort(a)
```

```python
# İki boyutlu arraylerde sıralama
m=np.random.randint(3,75,(2,3))

# Sütun bazında sıralama yapar(axis=0)
print(np.sort(m,axis=0))
print()

# Satır bazında sıralama yapar(axis=1)
print(np.sort(m,axis=1))
```

```python
# argsort() metodu
a=np.random.randint(5,82,(1,10))
print(a)
print()

# Sıralı dizinin indekslerini döndürür
print(np.argsort(a))
```

###Koşullu Eleman İşlemleri

NumPy dizilerinde belirli bir koşulu sağlayan elemanları seçmek, değiştirmek veya filtrelemek için kullanılır.

```python
a = np.arange(2,10)
print(a)

kosul=a%2==0
print("Çift sayılar:",a[kosul])
kosul=a>5
print("5'ten büyük sayılar:",a[kosul])
kosul=a!=7
print("7 den farklı sayılar:",a[kosul])
kosul=(a>3) & (a<7)
print("3'ten büyük 7'den küçük sayılar:",a[kosul])
```

```python
# np.where() ile Koşullu Seçim
result = np.where(a > 5, "Büyük", "Küçük")
print(result)
```

###Matematiksel İşlemler

- !!!Numpy arrayleri bir değişken gibi davranır. Üzerinde doğrudan matematiksel işlem gerçekleştirilebilir.

```python
x=np.arange(1,25,3)
y=np.arange(5,28,3)

print(x)
print(y)
```

```python
# Diziler doğrudan değişken gibi davranır
z= x+y
print("Toplama:",z)
z= x-y
print("Çıkarma:",z)
z= x*y
print("Çarpma:",z)
z= x**3
print("Küp alma:",z)
z= np.log(x)
print("Log alma:",z)
z= np.sin(x)
print("Sin değeri:",z)
```

**❗Metot kullanarak gerçekleşen işlemler**

```python
print("Toplama:",np.add(x,y))
print("Çıkarma:",np.subtract(x,y))
print("Çarpma:",np.multiply(x,y))
print("Bölme:",np.divide(x,y))
print("Üs alma:",np.power(x,3))
print("Log alma:",np.log10(x))
print("Sin alma:",np.sin(x))
print("Mod alma:",np.mod(x,2))
print("Mutlak değer:",np.absolute(x-y))
```

###İstatiksel İşlemler

NumPy dizileri üzerinde çeşitli istatistiksel hesaplamalar yapmayı sağlar.

**Temel İstatistiksel Fonksiyonlar:**

np.mean(arr) → Ortalamayı hesaplar.

np.median(arr) → Medyanı (ortadaki değeri) hesaplar.

np.std(arr) → Standart sapmayı hesaplar.

np.var(arr) → Varyansı hesaplar.

```python
m=np.random.randint(0,112,(3,3))
print(m)
```

**Pandas Series (Seriler)**, Pandas kütüphanesinde tek boyutlu, dizinlenmiş (indexed) veri yapılarıdır. Liste, NumPy array veya sözlükten oluşturulabilir ve farklı veri tiplerini içerebilir.

```python
# Kütüphanenin import edilmesi
import pandas as pd
```

###Pandas ile Farklı Dosya Tiplerinden Veri Okuma

Pandas kütüphanesi, birçok farklı dosya formatından veri okumayı destekler.

- **CSV Dosyası Okuma (read_csv):**
CSV, verilerin virgülle ayrılmış metin formatında saklandığı yaygın bir dosya türüdür ve pd.read_csv("dosya.csv") ile kolayca okunabilir.

- **Excel Dosyası Okuma (read_excel):**
Excel dosyaları, tabloları içeren ve sayfalara ayrılmış verileri saklamak için kullanılır; pd.read_excel("dosya.xlsx") ile belirli bir sayfa okunabilir.

- **JSON Dosyası Okuma (read_json):**
JSON, verileri anahtar-değer çiftleri ve hiyerarşik yapılarla saklayan bir format olup, pd.read_json("dosya.json") ile kolayca veri alınabilir.

- **Parquet Dosyası Okuma (read_parquet):**
Parquet, büyük veri işleme uygulamaları için optimize edilmiş ve sıkıştırılmış bir format olup, pd.read_parquet("dosya.parquet") ile okunabilir.

- **SQL Veritabanından Veri Okuma (read_sql):**
SQL veritabanları, yapısal veriyi saklamak için kullanılır ve pd.read_sql("SELECT * FROM tablo_adi", conn) komutuyla sorgular çalıştırılarak veri alınabilir.

- **HDF5 Dosyası Okuma (read_hdf):**
HDF5, büyük veri setlerini hiyerarşik olarak saklamak için kullanılan bir formattır ve pd.read_hdf("dosya.h5", key="veri_anahtari") ile veri erişilebilir.

- **Pickle Dosyası Okuma (read_pickle):**
Pickle, Python nesnelerini serileştirip saklamak için kullanılır ve pd.read_pickle("dosya.pkl") ile daha önce kaydedilen nesneler tekrar yüklenebilir.

### DataFrame'de Filtreleme İşlemleri

Pandas DataFrame'de filtreleme işlemi, belirli koşullara uyan satırları seçmek için kullanılır.

- Koşula Göre Filtreleme

- Birden Fazla Koşul ile Filtreleme (&,| operatörü)

- query() ile Filtreleme

**Koşula Göre Filtreleme**

```python
data=np.random.randint(10,100,(15,5))
df=pd.DataFrame(data,columns=["col1","col2","col3","col4","col5"])
print(df)
```

**Birden Fazla Koşul ile Filtreleme (&,| operatörü)**

```python
result=df.query("col1>=50 | col1<=70")[["col1","col2"]]
print(result)
```

### DataFrame Groupby

**groupby() fonksiyonu**, Pandas kütüphanesinde veri çerçevelerini belirli bir sütuna göre gruplamak ve bu gruplar üzerinde çeşitli işlemler yapmak için kullanılır.

**Özellikleri:**

- Bir DataFrame'i belirli bir sütuna göre gruplar.

- Her grup üzerinde toplama (sum()), ortalama (mean()), maksimum (max()), minimum (min()) gibi işlemler yapılabilir.

```python
personeller={
    "Çalışan":["Ahmet","Mehmet","Ali","Ayşe","Can"],
    "Departman":["İnsan Kaynakları","Bilgi İşlem","İnsan Kaynakları","Bilgi İşlem","İnsan Kaynakları"],
    "Yas":[30,25,45,50,23],
    "Semt":["Kadıköy","Tuzla","Maltepe","Tuzla","Kadıköy"],
    "Maaş":[3000,3500,2500,4000,3500]
}
```

```python
df=pd.DataFrame(personeller)
print(df)
```

```python
result=df.groupby("Departman")
# Tek bir grup get_group() metoduyla alınır.
print(result.get_group("Bilgi İşlem").sum())
```

```python
result=df.groupby("Semt").get_group("Tuzla")  # Semti Tuzla olanları getirir
print(result)
```

```python
result=df.groupby("Departman")[["Yas","Maaş"]].sum() # Departmana göre yaş ve maaş ortalaması
print(result)
```

```python
result=df.groupby("Departman")["Maaş"].mean() # Departmana göre maaş ortalamaları
print(result)
```

```python
result=df.groupby("Semt")["Çalışan"].count() # Semte göre çalışan sayıları
print(result)
```

❗agg() (aggregate) metodu, Pandas'ta groupby() ile birlikte kullanılarak birden fazla istatistiksel işlemi aynı anda uygulamaya yarar.

```python
result=df.groupby("Departman")["Maaş"].agg(['sum','mean','max','min'])
print(result)
```

### Pandas ile Kayıp ve Bozuk Veri Analizi

Pandas ile kayıp ve bozuk veri analizi, veri ön işleme aşamasında sıkça karşılaşılan bir işlemdir. Veri setlerinde eksik değerler (NaN) ve hatalı veriler (örneğin, yanlış formatta veya mantıksız değerler) olabilir. Bu verileri düzgün bir şekilde analiz etmek, temizlemek ve düzeltmek oldukça önemlidir.

- Pandas, eksik verileri NaN (Not a Number) ile temsil eder.

**Analiz Yöntemleri:**

 - Kayıp verileri tespit etmek için isnull() ve notnull() kullanılır. isnull(),NaN olan hücreleri True, olmayanları False döndürür.notnull(),NaN olmayan hücreleri True, NaN olanları False döndürür

 - Kayıp verileri temizlemek için dropna() (silme) veya fillna() (doldurma) kullanılır.

 - Bozuk veriler için veri türü dönüşümleri ve mantıksız değerlerin düzeltilmesi gerekir.

```python
import pandas as pd
import numpy as np
```

```python
data=np.random.randint(5,78,(5,3))
df=pd.DataFrame(data,index=['a','b','c','d','e'],columns=["column1","column2","column3"])
print(df)
```

```python
df=df.reindex(['a','b','c','d','e','f','g','h'])  # reindex() ile önceden olmayan satırlar eklenir ve bilgileri NaN olarak atanır.
print(df)
```

```python
result=df.drop("column1",axis=1) # Sütun bazında silme
print(result)
print()
result=df.drop("a",axis=0) # Satır bazında silme
print(result)
```

```python
# isnull() kullanımı(NaN olan hücreleri True, olmayanları False döndürür)
result=df.isnull()
print(result)
```

```python
result=df.isnull().sum()  # Sütunlardaki toplam boş alan sayısı
print(result)
```

```python
# Yeni sütun ekleme
newColumn=[np.nan,30,np.nan,51,np.nan,30,np.nan,10]
df["column4"]=newColumn
print(df)
```

```python
# Koşullu eleman işlemleri
result=df[df["column4"].notnull()]  # NaN olmayan değerleri getirir
print(result)
```

```python
result=df[df["column1"].isnull()]  # column1 sütununda NaN olan değerleri getirir
print(result)
```

**❕dropna() metodu**

Bu metot, Pandas'ta eksik (NaN) verilere sahip satırları veya sütunları veri çerçevesinden kaldırmak için kullanılır.

**Özellikler:**

 - axis=0: Satırları siler (varsayılan davranış).
 - axis=1: Sütunları siler.
 - how='any': Eğer bir satırda veya sütunda en az bir NaN varsa, o satırı/sütunu siler (varsayılan).
 - how='all': Eğer bir satırda veya sütunda tamamen NaN varsa, o satırı/sütunu siler.
 - thresh=N: Bir satır veya sütunun silinmemesi için en az N adet non-NaN değerinin olması gerektiğini belirtir.

```python
result=df.dropna() # NaN değer içeren satır veya sütunu siler(axis'in ön tanımlı değeri 0 olduğundan doğrudan kullanımda satırı siler)
print(result)
```

```python
result=df.dropna(axis=1) # Sütun bazında silme işlemini gerçekleştirir.
print(result)
```

```python
result=df.dropna(how="all") # Tüm değerleri NaN olan satırları siler(axis=0)
print(result)
```

```python
result=df.dropna(how="any") # Satırdaki/Sütundaki 1 değer bile NaN ise o satırı veya sütunu siler(axis=0 olduğu için satır sildi)
print(result)
```

```python
result=df.dropna(subset=["column1","column2"],how="all") # Özel olarak bakmak istenilen sütunları belirlemek için subset kullanılır.
print(result)
```

```python
result=df.dropna(thresh=2) # En az 2 tane normal veri varsa o satır/sütunu silmez!(axis=0 (satır))
print(result)
```

**❕fillna() metodu**

Bu metot, Pandas'ta eksik (NaN) verilere sahip hücreleri belirli bir değerle doldurmak için kullanılır. Kayıp verilerin yerine mantıklı bir değer koyarak veri setini temizlemenizi sağlar.

**Özellikler:**

 - value: Kayıp değerlerin hangi değerle doldurulacağını belirler. Tek bir değer ya da her sütun için farklı bir değer olabilir.

 - method: Doldurma işleminin yöntemini belirtir (örneğin, 'ffill' veya 'bfill').
   - 'ffill': "Forward fill" (önceki değeri takip ederek doldurur).
   - 'bfill': "Backward fill" (sonraki değeri takip ederek doldurur).
   
 - axis: Hangi eksende (satır veya sütun) işlem yapılacağını belirtir. (Varsayılan: axis=0 — satırlar için)

 - inplace: Eğer True olarak ayarlanırsa, değişiklikler orijinal DataFrame üzerinde yapılır; yoksa yeni bir DataFrame döndürülür.

```python
result=df.fillna(value="No Value") # NaN değerleri No Value ile doldurur
print(result)
```

```python
result=df.fillna(value=1) # NaN değerleri 1 ile doldurur.
print(result)
```

```python
# DataFrame'de NaN değerleri ortalama değerle değiştirme
mean_df=df.mean()
result=df.fillna(value=mean_df)
print(result)
```

```python
result=df.fillna(df.median())  # NaN değerleri medyan ile doldurma
print(result)
```

**ffill()-bfill() metotlarının kullanımı**

```python
result=df.fillna(df.ffill()) # NaN değerleri önceki değerler ile doldurma
print(result)
```

```python
result=df.fillna(df.bfill()) # NaN değerleri sonraki değerler ile doldurma
print(result)
```

### DataFrame Birleştirme

Pandas’ta DataFrame birleştirme işlemleri, farklı veri setlerini tek bir yapıda toplamak için kullanılan yöntemleri kapsar.

**Pandas’ta üç temel DataFrame birleştirme yöntemi vardır:**

- merge() → SQL tarzı birleştirme (ortak sütuna göre)
- join() → İndeks bazlı birleştirme
- concat() → DataFrame'leri alt alta veya yan yana ekleme

```python
customers={
    'CustomerID':[1,2,3,4],
    'FirstName':["Ahmet","Ali","Hasan","Canan"],
    'LastName':["Yılmaz","Korkmaz","Çelik","Toprak"]
}

orders={
    'OrderID':[10,11,12,13],
    'CustomerID':[1,2,5,7],
    'OrderDate':["2010","2010","2012","2013"]
}
```

```python
df_customers=pd.DataFrame(customers)
df_orders=pd.DataFrame(orders)
```

**❕merge(): SQL Benzeri Birleştirme**

İki DataFrame’i belirli bir sütuna göre birleştirir.

- on="sütun_adı" ile eşleşme yapılır.

**how parametresi ile birleşim türü belirlenir:**
 - "inner" → Sadece ortak olanları alır.
 - "left" → Sol tabloyu korur, sağdaki eksik değerler NaN olur.
 - "right" → Sağ tabloyu korur, soldaki eksik değerler NaN olur.
 - "outer" → Tüm verileri alır (eksik yerler NaN olur).

```python
result=pd.merge(df_customers,df_orders,how="inner")
print(result)
```

```python
result=pd.merge(df_customers,df_orders,how="left") # Sol tabloyu korur
print(result)
```

```python
result=pd.merge(df_customers,df_orders,how="right") # Sağ tabloyu korur
print(result)
```

```python
result=pd.merge(df_customers,df_orders,how="outer") # Tüm veriyi alır
print(result)
```

**❕join(): İndeks Bazlı Birleştirme**

- İndekslere göre birleşim yapar.
- df1 ve df2'de ortak indeksleri kullanır.
lsuffix ve rsuffix parametreleriyle aynı sütun adlarının çakışması önlenir.

```python
result=df_customers.join(df_orders,lsuffix="_left",rsuffix="_right",how="inner")
print(result)
```

```python
result=df_customers.join(df_orders,lsuffix="_left",rsuffix="_right",how="outer") # İki tabloyuda birleştirir
print(result)
```

**❕concat(): DataFrame'leri Alt Alta veya Yan Yana Eklemek**

- Dikey (axis=0) veya yatay (axis=1) ekleme yapar.
- ignore_index=True ile indeksleri sıfırlanabilir.(index problemini çözmek adına ignore_index parametresi True yapılmalı)

```python
customerA={
    'CustomerID':[1,2,3,4],
    'FirstName':["Ahmet","Ali","Hasan","Canan"],
    'LastName':["Yılmaz","Korkmaz","Çelik","Toprak"]
}

customerB={
    'CustomerID':[4,5,6,7],
    'FirstName':["Yağmur","Çınar","Cengiz","Can"],
    'LastName':["Bilge","Turan","Turan","Yılmaz"]
}
```

```python
df_customerA=pd.DataFrame(customerA)
df_customerB=pd.DataFrame(customerB)
```

```python
result=pd.concat([df_customerA,df_customerB],ignore_index=True,axis=0) # Alt alta birleştirme(axis=0 satır bazlı birleştirme!)
print(result)
```

```python
result=pd.concat([df_customerA,df_customerB],ignore_index=True,axis=1) # Yan yana birleştirme(axis=1 sütun bazlı birleştirme!)
print(result)
```

```python
# Farklı bir durum
df_customerA.columns=("CustomerID","FirstName","column1")  # df'te sütun ismini değiştirerek ortak olmayan veriler sağlandı

# Yanlış Kullanım
# pd.concat([df_customerA,df_customerB],ignore_index=True) HATALI KULLANIM! Ortak olmayan sütunlar olduğundan kesişim üzerinden birleştirme yapılmalı

# Doğru Kullanım
result=pd.concat([df_customerA,df_customerB],ignore_index=True,join="inner")
print(result)
```

**❗DataFrame'de Sıralama İşlemi**

Pandas’ta DataFrame sıralama (sorting) işlemi, verileri belirli bir sütuna göre artan veya azalan şekilde düzenlemeyi sağlar. İki temel sıralama yöntemi vardır:

- sort_values() → Belirli bir sütuna göre sıralama
- sort_index() → İndekse göre sıralama

**sort_values(): Sütuna Göre Sıralama**

Bir veya birden fazla sütuna göre sıralama yapar.

- ascending=True → Artan sıralama (varsayılan).
- ascending=False → Azalan sıralama.
